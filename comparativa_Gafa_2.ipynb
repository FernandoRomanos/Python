```python
# Paso 2: importar las librerias

import yfinance as yf
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Paso 3: Comparar cotizaciones
# Definir los símbolos de las acciones
stocks = ['GOOGL', 'AAPL', 'MSFT', 'META', 'AMZN']
```

```python
from typing_extensions import dataclass_transform

# Descargar las acciones
data = yf.download(stocks, start='2013-01-01', end='2023-05-31')['Adj Close']
```

```python
# Paso 3: graficos
from tempfile import template
```

```python
# grafico de lineas
fig_line = go.Figure()
for stock in stocks:
    fig_line.add_trace(go.Scatter(x=data.index, y=data[stock], name=stock))

fig_line.update_layout(
    title='Cotizaciones de las acciones',
    xaxis_title='Fecha',
    yaxis_title='Precio',
    template='plotly_dark',
    hovermode='x unified'  # Muestra los tooltips de todas las trazas al mismo tiempo
)
fig_line.show()
```
