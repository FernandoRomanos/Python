# Paso 2: importar las librerias

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# Paso 3: Obtener los datos

```python
# Símbolos de las cotizaciones
symbols = {
    'Petróleo': 'CL=F',
    'Gas Natural': 'NG=F',
    'Oro': 'GC=F'
}
```

```python
# Descarga de datos históricos
data = {}
for name, symbol in symbols.items():
    data[name] = yf.download(symbol, start='2014-01-01', end='2023-12-31')['Adj Close']
```

```python
# Combinar en un solo DataFrame
df = pd.DataFrame(data)
```

# Paso 4: Procesar y visualizar los datos

```python
plt.figure(figsize=(14, 7))
```

```python
for colum in df.columns:
    plt.plot(df.index, df[colum], label=colum)
plt.title('Comparación de cotizaciones: Petróleo, Gas Natural y Oro')
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
