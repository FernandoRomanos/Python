```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Paso 2: Obtener datos
# Símbolos de los índices bursátiles en Yahoo Finance
index_symbols = {
    'S&P 500 (Nueva York)': '^GSPC',
    'FTSE 100 (Londres)': '^FTSE',
    'Hang Seng Index (Hong Kong)': '^HSI',
    'Straits Times Index (Singapur)': '^STI',
    'IBEX 35 (Madrid)': '^IBEX'
}
```

```python
# Descarga de datos históricos
index_data = {}
for name, symbol in index_symbols.items():
    index_data[name] = yf.download(symbol, start='2014-01-01', end='2024-12-31')['Adj Close']
```

```python
# Combinar en un solo DataFrame y elegir filas y columnas con valores NaN
df_indices = pd.DataFrame(index_data).dropna()
```

```python
# Paso 3: Procesar y visualizar los datos
plt.figure(figsize=(14, 7))
```

```python
for column in df_indices.columns:
    plt.plot(df_indices.index, df_indices[column], label=column)
plt.title('Comparación de cotizaciones de índices bursátiles')
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 7))

corr_matrix = df_indices.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mapa de calor de la correlación entre índices bursátiles')
plt.show()
```

```python
plt.figure(figsize=(14, 7))

df_indices.plot(kind='box')
plt.title('Distribución de cotizaciones de índices bursátiles')
plt.xlabel('Índice')
plt.ylabel('Precio (USD)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))

for column in df_indices.columns:
    plt.fill_between(df_indices.index, df_indices[column], alpha=0.5, label=column)

plt.title('Comparación de cotizaciones de índices bursátiles')
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
