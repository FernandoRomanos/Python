# Comparativa de las cotizaciones de los últimos 10 años de: Apple, Microsoft, Google, Amazon y Meta

## Utilizamos matplotlib

```python
# Paso 1:Importar bibliotecas necesarias
!pip install yfinance matplotlib
```

```python
# Paso 2: importar las librerias

import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Paso 3: Comparar cotizaciones
# Definir los símbolos de las acciones
stocks = ['GOOGL', 'AAPL', 'MSFT', 'META', 'AMZN']
```

```python
from typing_extensions import dataclass_transform
# Descargar las acciones
data = yf.download(stocks, start='2013-01-01', end='2023-05-31')['Adj Close']
```

```python
# Paso 4: Realizar los graficos
plt.figure(figsize=(14, 7))
for stock in stocks:
    plt.plot(data[stock], label=stock)
plt.title('Cotizaciones de las acciones')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.grid(True)
plt.show
```
