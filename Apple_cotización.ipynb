```python
import yfinance as yf
import plotly.graph_objects as go
```

```python
# Paso 3: Obtener los datos históricos de Apple
apple_stock = yf.Ticker("AAPL")
hist = apple_stock.history(period="max")
```

```python
# Paso 4: Crear el gráfico de líneas
fig = go.Figure()
```

```python
# Agregar la línea de tiempo de los precios de cierre
fig.add_trace(go.Scatter(x=hist.index, y=hist['Close'], mode='lines', name='AAPL Close'))
# Personalizar el diseño del gráfico
fig.update_layout(
    title='Cotización Histórica de Apple',
    xaxis_title='Fecha',
    yaxis_title='Precio de Cierre (USD)',
    xaxis_rangeslider_visible=True
)

# Mostrar el gráfico
fig.show()
```
